```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csv_url = 'https://raw.githubusercontent.com/zaladmander/Analyzing-Attributes-of-AI-Generated-vs.-Human-Written-Text/refs/heads/main/ai_vs_human_dataset_medium.csv'
df = pd.read_csv(csv_url)
```

```python
print(df.columns)
```

```python
print("\nFirst 5 records:\n", df.head(5))

print("\nDataset shape:", df.shape)

print("\nData description: \n", df.describe())

plt.figure(figsize=(12,5))
sns.boxplot(data=df, x='label', y='plagiarism_score', palette='Set1')
plt.title("Plagiarism Score by Label")
plt.show()
```
